## 2. Data Preprocessing

```python
# Loads a DataFrame with the shape of "stroke". Each element is either True if it's a missing value (NaN) and False if otherwise.
# It would also be possible to use isna(), it's an alias and creates the same outcome.

stroke.isnull()
```

```python
# isna() checks the DataFrame for missing values (NaN). 
# any() check if there is atleast one True value. By Default it checks for True values along the axis 0.
# Chained with any() the second any() checks if there is a True value returned by the first any(). It checks if there is atleast one missing value in the DataFrame.
if stroke.isna().any().any():
    print("DataFrame contains missing values.")
else:
    print("DataFrame does not contain any missing values.")
```

```python
# .sum(axis=0) sums the number of True values along the axis 0. It's summing the number of missing values in each column.
# .sum sums the values in the series returned by .sum(axis=0). It calculates the total amount of missing values in the DataFrame for each column.

total_missing_values = stroke.isna().sum(axis=0).sum()
print("Total count of missing values in the DataFrame:", total_missing_values)
```

```python
# stroke.select_dtypes(include=['number']) This only selects columns in stroke that has numeric values. It then returns a DataFrame only containing numeric columns.
# stroke.select_dtypes(exclude=['number']) this only selects columns in stroke that has non-numeric values. It the returns a DataFrame only containing non-numerice columns.

missing_values_numeric = stroke.select_dtypes(include=['number']).isna().sum().sum()
missing_values_non_numeric = stroke.select_dtypes(exclude=['number']).isna().sum().sum()

print("Total count of missing numeric values in the DataFrame:", missing_values_numeric)
print("Total count of missing non-numeric values in the DataFrame:", missing_values_non_numeric)
```

```python
# Since ID dosen't make any sense for this DataFrame I have chosen to remove it.

stroke.drop('id',axis=1, inplace=True)
```

```python
# Checking the DataFrame for any duplicates

stroke.duplicated().any()
```
